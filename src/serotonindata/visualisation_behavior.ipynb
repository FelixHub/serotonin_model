```python
import numpy as np
from behavior_processing import GainChangeEvents
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python

data_behaviour = {}
for mouse in mice :
    # mouse = {"prefix" : "RL2" , "mouse_id" : 1}
    print(mouse)
    mouse_id = mouse["mouse_id"]
    RL_id = mouse['prefix']
    uid =  RL_id+"_"+str(mouse_id)
    key = {"prefix" : RL_id, "mouse_id":mouse_id}
    session_keys = (
        (GainChangeEvents() & key).fetch("KEY")
    )

    for i_session, session_key in tqdm(enumerate(session_keys)):

        sample_times, df_over_f = (ProcessedPhotometry() & session_key ).fetch1(
            "sample_times","df_over_f"
        )
        running_speed,gain,trial_num,wheel_position= (PhotometrySyncBehavior() & session_key ).fetch1(
                "running_speed",
                "gain",
                "trial_num",
                "wheel_position"
            )

        data_behaviour[uid+'_'+str(i_session)] =  {k:v for k,v in zip(["sample_times","df_over_f","running_speed","gain","trial_num","wheel_position"],
                                                    [sample_times,df_over_f,running_speed,gain,trial_num,wheel_position])}

        # data.to_csv('mice_data_'+uid+'.csv')
```
